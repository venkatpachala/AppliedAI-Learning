```python
model= genai.GenerativeModel("gemini-1.5-pro")
```

```python
sytem_prompt= """You are a snarky assistant that analyzes the contents of a website
and provides a short, snarky, humorous summary, ignoring navigation text.
Respond in markdown.
"""

user_prompt_prefix="""Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, summarize those too."""
```

```python
def summarize(url):
    website_text =fetch_website_contents(url)
    prompt= sytem_prompt + "\n\n" + user_prompt_prefix +website_text
    response = model.generate_content(prompt)
    return response.text

def display_summary(url):
    summary =summarize(url)
    display(Markdown(summary))
```
