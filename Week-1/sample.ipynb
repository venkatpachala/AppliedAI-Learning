```python
import time
import os
from dotenv import load_dotenv
from google import genai

load_dotenv()

client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])

MODEL_FOR = "models/gemini-pro-latest"
MODEL_AGAINST = "models/gemini-flash-latest"


def call_gemini(model, prompt):
    response = client.models.generate_content(
        model=model,
        contents=prompt,
        config={
            "max_output_tokens": 120,
            "temperature": 0.7
        }
    )
    return response.text.strip()


def debate(topic, rounds=6):
    print(f"\nðŸ”¥ DEBATE STARTED: {topic}\n")

    history = []

    for turn in range(rounds):
        stance = "FOR" if turn % 2 == 0 else "AGAINST"
        model = MODEL_FOR if stance == "FOR" else MODEL_AGAINST

        context = "\n".join(history[-2:])

        prompt = f"""
Topic: {topic}

Your stance: {stance}

Rules:
- Make ONE strong argument
- Max 2 sentences
- Do NOT agree with the opposite stance
- Do NOT repeat earlier points

Previous arguments:
{context}

Answer:
"""

        reply = call_gemini(model, prompt)
        history.append(reply)

        print(f"{stance} ({model}):")
        print(reply)
        print("-" * 60)

        time.sleep(1)

    print("\nðŸ›‘ DEBATE ENDED\n")
```

```python
debate("Should governments heavily regulate artificial intelligence?")
```
