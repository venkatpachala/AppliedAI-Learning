```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
google_api_key=os.getenv("GOOGLE_API_KEY")
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

if google_api_key:
    print(f"Your google api exists and begin with {google_api_key[:2]}")
else:
    print("Google API key not set")

```

```python
# openai = OpenAI()
#This is connecting OpenAI cleint library
#A thin wraper around calls to HTTP endpoints

gemini_url="https://generativelanguage.googleapis.com/v1beta/openai/"
#Here we use OpenAI python client to call

gemini=OpenAI(api_key=google_api_key, base_url=gemini_url)
openrouter_url = "https://openrouter.ai/api/v1"
```

```python
question=[
    {"role": "user", "content": "Tell me a joke on PM MODI"},
]
```

```python
response=gemini.chat.completions.create(model="gemini-2.5-flash",messages=question)
display(Markdown(response.choices[0].message.content))
```
